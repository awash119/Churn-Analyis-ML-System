### 1.Import Dependencies

```python

import numpy as np 
import pandas as pd 
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv("data\\processed\\ChurnModelling_Feature_Encoding_Applied.csv")
df
```

```python
features_needed_to_scale=["Age","Tenure","Balance","EstimatedSalary"]
for col in features_needed_to_scale:
    scaler=StandardScaler()
    df[col]=scaler.fit_transform(df[col])
df
```

```python
df.to_csv("data\\processed\\ChurnModelling_Final.csv",index=False)   
```

```python
columns_need_to_be_scaled = ['Age', 'Tenure', 'Balance', 'Estimated Salary']

for col in columns_need_to_be_scaled:
    scalar = StandardScaler()

    df[col] = scalar.fit_transform(df[[col]])    #Method 1
    df[col] = scalar.fit_transform(df[col].values.reshape(10000,1)) # Method 2 But Risky We Needed to Know the number of rows in advance
    df[col] = scalar.fit_transform(df[col].values.reshape(-1,1))    # Method 3 Recommended


df
```
