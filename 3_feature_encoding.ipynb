### 1.Import Dependencies

```python

import numpy as np 
import pandas as pd 
```

```python
df=pd.read_csv("data\processed\ChurnModelling_Binning_Applied.csv",)
df
```

### 2.Encode Nomial Vairable

```python
gender_dummies=pd.get_dummies(df['Gender'],prefix='Gender')
credit_score_dummies=pd.get_dummies(df['Geography'],prefix='Geography')
df_encoded=pd.concat([df,credit_score_dummies,gender_dummies],axis=1)
df_encoded.drop(columns=['Geography','Gender'],inplace=True)
df_encoded

```

### 2.Encode Ordinal Variable

```python
encode_dict_creditscore = {
    'Poor': 0,
    'Fair': 1,
    'Good': 2,
    'Very Good': 3,
    'Excellent': 4
}
df_encoded['CreditScoreBins'] = df_encoded['CreditScoreBins'].map(encode_dict_creditscore)
df_encoded
```
